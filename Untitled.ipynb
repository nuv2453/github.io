```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.datasets import make_blobs

# Generate synthetic data
np.random.seed(42)
X, y = make_blobs(n_samples=300, centers=4, random_state=42, cluster_std=1.0)

# Create a Gaussian Mixture Model with 4 components
gmm = GaussianMixture(n_components=4, random_state=42)

# Fit the model to the data
gmm.fit(X)

# Predict the clusters
labels = gmm.predict(X)

# Plot the data points and cluster assignments
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50, edgecolors='k', linewidths=1)
plt.scatter(gmm.means_[:, 0], gmm.means_[:, 1], color='red', marker='x', s=200, linewidths=2, label='GMM Centers')
plt.title('Gaussian Mixture Model')
plt.legend()
plt.show()
```
